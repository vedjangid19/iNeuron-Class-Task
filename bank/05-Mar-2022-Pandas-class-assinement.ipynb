#### Bank DataSet Que

```python
import pandas as pd
data = pd.read_csv('bank.csv' , sep=";")
data
```

```python
data.columns
```

```python
#QUE. 1 . how many campaign available in this dataset . 
data.campaign.nunique()

#QUE. 2 . how many users do we have with housing and personal loan . 
data[(data.housing == "yes") & (data.loan == "yes")]['housing'].count()

#QUE. 3. how many person do we have whose age is 60+ . 
data[data.age > 60].age.count()

#QUE. 4 . in which month we have trageted most of the customer . 
data.groupby("month")["duration"].sum().idxmax()

#QUE. 5 . which mode of call is giving you more result
data[data.y == "yes"].groupby(['contact'])["y"].count().idxmax()

#QUE. 6 . how many enterpeures do we have in this list  
data[data.job == "entrepreneur"].job.count()

#QUE. 7 . how many customers do we have with negative balance 
data[data.balance < 0].balance.count()

#QUE. 8. prepare a group of data based on education level . 
agg_fun = dict(age='mean', job='count', balance='mean', housing='count', loan='count', duration='mean',campaign='count')
data.groupby("education").agg(agg_fun)
```

#### Titanic DataSet Que

```python
# For titanic datset : 
import pandas as pd
data = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
print(data.columns)

#QUE. 1 . Find out how many male and female passenger was onboarded .
data.groupby("Sex").PassengerId.count()

#QUE. 2 . how many survived we have .
data[data['Survived'] == 1]['PassengerId'].count()

#QUE. 3 . how many casuality we have.


#QUE. 4 . what is name of a person who is the eldest one . 
data[data.Age == max(data.Age)]['Name']

#QUE. 5 . how many passenger do we have in first , second and third class.
data.groupby('Pclass').count().PassengerId

#QUE. 6 . how many person we have whose name starts with "s"
len([x for x in data["Name"] if x.startswith("s")])

#QUE. 7 . try to create a new column which is a summation  of "SibSp" and "parch"
data['summation'] = data['SibSp'] + data['Parch']
data

#QUE. 8 . how many person do we have below age of 25 .
data[data.Age < 25]['Age'].count()

#QUE. 9 . how many person died whose age was less then 40 
data[(data.Survived == 0) & (data.Age < 40)]['PassengerId'].count()

#QUE. 10 . from a  cabin column seperate text and numeric value . 
data['number'] = data['Cabin'].str.extract('(\d+)')
data['text'] = data['Cabin'].str.extract('([a-zA-Z])')

data
```
